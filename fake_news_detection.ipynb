## Creating a function to process the texts

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
df["text"] = df["text"].apply(wordopt)
```

```python
df.head()
```

## Defining dependent and independent variables

```python
x = df["text"]
y = df["class"]
```

## Splitting Training and Testing

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

## Convert text to vectors

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
import requests
from bs4 import BeautifulSoup
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

class FakeNewsDetectionSystem:
    def __init__(self):
        self.vectorizer = CountVectorizer(stop_words='english')
        self.model = MultinomialNB()

    def preprocess_text(self, text):
        # Implement your text preprocessing here (e.g., cleaning, tokenization)
        return text

    def get_article_text(self, url):
        # Extract text from a news article using web scraping
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            article_text = " ".join([p.get_text() for p in soup.find_all('p')])
            return article_text
        else:
            return None

    def predict_fake_news(self, text):
        # Preprocess the text
        preprocessed_text = self.preprocess_text(text)

        # Convert text to numerical features using the same vectorizer used for training
        text_vec = self.vectorizer.transform([preprocessed_text])

        # Make a prediction using the loaded model
        prediction = self.model.predict(text_vec)
        return prediction[0]

    def train_model(self, training_data, labels):
        # Preprocess training data
        X_train = [self.preprocess_text(text) for text in training_data]
        X_train_vec = self.vectorizer.fit_transform(X_train)

        # Train the model
        self.model.fit(X_train_vec, labels)

    def run(self):
        # Sample training data (replace with your actual training data)
        training_data = [
            "Real news text 1",
            "Fake news text 1",
            # ... add more training data
        ]
        labels = [0, 1,  # ... add more labels
        ]

        # Train the model
        self.train_model(training_data, labels)

        while True:
            url = input("Enter the URL of a news article (or 'exit' to quit): ")

            if url.lower() == 'exit':
                break

            article_text = self.get_article_text(url)

            if article_text is None:
                print("Error: Unable to retrieve article text.")
            else:
                prediction = self.predict_fake_news(article_text)
                if prediction == 1:
                    print("This article is likely fake.")
                else:
                    print("This article appears to be real.")

if __name__ == "__main__":
    system = FakeNewsDetectionSystem()
    system.run()
```
